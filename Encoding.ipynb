# Working on sample data for testing and practicing Target and One Hot Encoding 
Cetgorical data one most essential information for any Data Analysis and ML problem.
But ML models cant recongnize Categorical Data. To convert Catogorical variables to Numerical variables we use Encoding.
Best and most used encoding technique is OH Encoding. 
Other most used ones are Label Encoding and Target Encoding

```python
import numpy as np
import pandas as pd

from category_encoders import TargetEncoder
```

```python
Animal = ['cat', 'hamster', 'cat', 'cat', 'dog', 'hamster', 'cat', 'dog', 'cat', 'dog']
Target = [1, 0, 0, 1, 1, 1, 0, 1, 0, 0]

# Zipping two list pairs elements from first list to second
data = list(zip(Animal,Target))
df = pd.DataFrame(data, columns=['Animal','Target'])
df.head(10)
```

```python
# Target Encoding
encoder = TargetEncoder()
df['Animal Encoded'] = encoder.fit_transform(df['Animal'], df['Target'])
df.head(10)
```

```python
# Label and One HOT Encoding 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

Animal = ['cat', 'hamster', 'cat', 'cat', 'dog', 'hamster', 'cat', 'dog', 'cat', 'dog']
Target = [1, 0, 0, 1, 1, 1, 0, 1, 0, 0]

# Zipping two list pairs elements from first list to second
data = list(zip(Animal,Target))
df = pd.DataFrame(data, columns=['Animal','Target'])

# Label Encoder
le = LabelEncoder()
df['Animal Encoded'] = le.fit_transform(df.Animal)
df.head(10)
```

```python
#One HOT Encoding 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

Animal = ['cat', 'hamster', 'cat', 'cat', 'dog', 'hamster', 'cat', 'dog', 'cat', 'dog']
Target = [1, 0, 0, 1, 1, 1, 0, 1, 0, 0]

# Zipping two list pairs elements from first list to second
data = list(zip(Animal,Target))
df = pd.DataFrame(data, columns=['Animal','Target'])

# Label Encoder
le = LabelEncoder()
df['Animal Encoded'] = le.fit_transform(df.Animal)
df.head(10)
# One 
encoder = OneHotEncoder(categories='auto')
X = encoder.fit_transform(df['Animal Encoded'].values.reshape(-1,1)).toarray()
dfonehot=pd.DataFrame(X)
df = pd.concat([df,dfonehot],axis=1)
df.columns = ['Animal', 'Target', 'Animal Encoded', 'isCat', 'isDog', 'IsHamster']

df.head(10)
```
